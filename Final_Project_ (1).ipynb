```python
import json
from sentence_transformers import SentenceTransformer

embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

with open("/content/transcription_segments.json", "r", encoding="utf-8") as f:
    segments = json.load(f)
```

```python
import chromadb

chroma_client = chromadb.PersistentClient(path="/mnt/data/chroma_db")

collection = chroma_client.get_or_create_collection(name="youtube_transcripts")
```

```python
for segment in segments:
    embedding = embedding_model.encode(segment["text"]).tolist()
    collection.add(
        ids=[str(segment["id"])],
        embeddings=[embedding],
        metadatas=[{
            "start": segment["start"],
            "end": segment["end"],
            "text": segment["text"]
        }]
    )

print("Embeddings stored in ChromaDB.")

```

```python
def search_chroma(query, top_k=3):
    query_embedding = embedding_model.encode([query]).tolist()
    results = collection.query(
        query_embeddings=query_embedding,
        n_results=top_k
    )

    return results

query = "How does AI impact jobs?"
results = search_chroma(query)

print("\n Segments:")
for i in range(len(results["documents"][0])):
    print(f"\n Segment {i+1}: ")
    print(f" Text: {results['metadatas'][0][i]['text']}")
    print(f" Time stamp: {results['metadatas'][0][i]['start']}s - Till: {results['metadatas'][0][i]['end']}s")
```

```python
def get_segments_by_time(start_time, end_time):
    """
    Retrieves transcript segments that fall within the given time range.
    """
    results = [
        seg for seg in segments
        if seg["start"] >= start_time and seg["end"] <= end_time
    ]

    return results

start_time = 10
end_time = 3600

segments_in_range = get_segments_by_time(start_time, end_time)

print(f"\n **Segments spoken between {start_time}s and {end_time}s:**")
for seg in segments_in_range:
    print(f" **Start:** {seg['start']}s - **End:** {seg['end']}s")
    print(f" **Text:** {seg['text']}\n")
```

```python
import chromadb
from sentence_transformers import SentenceTransformer
from langchain.chat_models import ChatOpenAI
from langchain.schema import Document
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceEmbeddings

embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

chroma_client = chromadb.PersistentClient(path="/mnt/data/chroma_db")
collection = chroma_client.get_or_create_collection(name="youtube_transcripts")

llm = ChatOpenAI(model="gpt-4", temperature=0.7, openai_api_key="OPENAI_API_KEY")

def retrieve_relevant_context(query, top_k=3):
    """
    Perform a similarity search in ChromaDB and return the most relevant transcript segments.
    """
    query_embedding = embedding_model.encode([query]).tolist()
    results = collection.query(
        query_embeddings=query_embedding,
        n_results=top_k
    )

    retrieved_docs = []
    for i in range(len(results["documents"][0])):
        retrieved_docs.append(
            Document(
                page_content=results["metadatas"][0][i]["text"],
                metadata={"start": results["metadatas"][0][i]["start"], "end": results["metadatas"][0][i]["end"]}
            )
        )

    return retrieved_docs

def generate_rag_response(query):
    """
    Uses retrieved transcript segments to generate an LLM-powered response.
    """
    retrieved_context = retrieve_relevant_context(query)

    context_text = "\n".join([doc.page_content for doc in retrieved_context])

    prompt = f"""
    You are an AI assistant with knowledge extracted from a YouTube transcript.
    Answer the user's question using the provided transcript context.

    Context:
    {context_text}

    Question: {query}
    Answer:
    """

    response = llm.predict(prompt)

    return response
```

```python
from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferMemory
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain.tools import Tool
from langchain.schema import Document
from langchain.agents import initialize_agent, AgentType
from pydantic import BaseModel, Field
import json
import gradio as gr

quiz_template = """
You are an AI that generates multiple-choice quiz questions.
Based on the following text, create {num_questions} well-structured multiple-choice questions.

Text: "{text}"

Format your response strictly as follows:
Question: <question text>
A) <option 1>
B) <option 2>
C) <option 3>
D) <option 4>
Correct Answer: <correct answer letter>
---
"""

quiz_prompt = ChatPromptTemplate.from_template(quiz_template)
memory = ConversationBufferMemory(return_messages=True, memory_key="history")
llm = ChatOpenAI(model_name="gpt-4o", temperature=0.5)

chatbot_chain = ConversationChain(
    llm=llm,
    memory=memory
)

class QuizzInput(BaseModel):
    text: str = Field(description="The extract to be passed to quizz"),
    num_questions: str = Field(description="Number of questions to be generated")


@tool('Quizz Generator', args_schema=QuizzInput, infer_schema=True )
def generate_quiz(text: str, num_questions: str) -> str:
    """Generates multiple-choice quiz questions based on the provided text and number of questions."""
    formatted_prompt = quiz_prompt.format(text=text, num_questions=num_questions)
    response = llm.invoke(formatted_prompt)
    return response

def generate_quiz_from_json(query):
    if isinstance(query, str):
        query = json.loads(query)

    if "text" in query and "num_questions" in query:
        return generate_quiz(query["text"], query["num_questions"])
    return "Invalid input format. Expected 'text' and 'num_questions'."

chatbot_tool = Tool(
    name="Chatbot",
    func=lambda query: chatbot_chain.predict(input=query),
    description="Generates responses based on conversation history and transcript."
)

def retrieve_relevant_context(query: str, top_k=3):
    query_embedding = embedding_model.encode([query]).tolist()
    results = collection.query(
        query_embeddings=query_embedding,
        n_results=top_k
    )

    retrieved_docs = [
        Document(
            page_content=results["metadatas"][0][i]["text"],
            metadata={
                "start": results["metadatas"][0][i]["start"],
                "end": results["metadatas"][0][i]["end"]
            }
        )
        for i in range(len(results["documents"][0]))
    ]

    return retrieved_docs

class RetrieverInput(BaseModel):
    query: str = Field(description="Query to look up in retriever")

@tool('YouTube_RAG', args_schema=RetrieverInput, infer_schema=True)
def generate_rag_response(query: str) -> str:
    """
    Uses retrieved transcript segments to generate an AI-powered response.
    """
    retrieved_context = retrieve_relevant_context(query)
    context_text = "\n".join([doc.page_content for doc in retrieved_context])

    prompt = f"""
    You are an AI assistant with knowledge extracted from a YouTube transcript.
    Answer the user's question using the provided transcript context.

    Context:
    {context_text}

    Question: {query}
    Answer:
    """

    response = llm.predict(prompt)
    return response

agent_prefix = """
You are EduAI, a friendly and enthusiastic AI learning assistant specializing in artificial intelligence education.
You have access to a database of YouTube video content about AI and can generate engaging quizzes to help learners test their knowledge.

Core Functions:
- Greet users warmly and show enthusiasm for helping them learn.
- Accept natural language queries about AI topics.
- Search through the RAG-enhanced database of YouTube video content.
- Provide relevant timestamps and summaries of video segments.
- Offer context and explanations in an accessible, conversational tone.
- Include direct quotes from videos when appropriate.
- Provide clear function execution (query or quiz).
- Offer additional help or clarification when needed.
- Always maintain a positive, solution-focused attitude.

Error Handling:
- If no relevant video content is found, acknowledge this warmly and suggest related topics.
- If quiz generation fails, offer alternative learning approaches.


If the input asks for a quizz, YOU MUST pass the quizz to the user. Otherwise, he won't be able to pass their class.
If the user does not ask for a quizz, query the database to find answer to their reponse. If you cannot provide one, say that you do not have enough information.

Here is your input {{input}}
"""

tools = [generate_rag_response, generate_quiz]

agent = initialize_agent(
    tools=[generate_rag_response, generate_quiz],
    llm=llm,
    agent=AgentType.STRUCTURED_CHAT_ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True,
    memory=memory,
    handle_parsing_errors=True,
    agent_kwargs={"prefix": agent_prefix}
)

def chatbot(query):
    try:
        response = agent.invoke(query)
    except:
        return 'The Agent is not able to reply to your quest right now'

    return response['output']
```
